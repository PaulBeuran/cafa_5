```python
import json
import pandas as pd

from cafa_5.dataload import read_prot_seq_from_fasta_as_df
```

```python
def read_amino_acids_info_from_txt_path(
    amino_acids_info_txt_path : str
) -> pd.DataFrame:
    
    amino_acids_info_df = pd.read_table(amino_acids_info_txt_path, sep=" ")
    return amino_acids_info_df
```

```python
amino_acids_info_df = read_amino_acids_info_from_txt_path(
    "../kaggle/input/amino_acids_info/amino_acids_info.txt"
)
amino_acids_info_df.set_index("amino_acid", inplace=True)

prot_seq_df = read_prot_seq_from_fasta_as_df("../kaggle/input/cafa-5-protein-function-prediction/Train/train_sequences.fasta")
prot_seqs = prot_seq_df["sequence"].tolist()

display(amino_acids_info_df, prot_seqs)
```

```python
type_seqs = []
is_hydroxylic_seqs = []
contains_sulphur_seqs = []
amidic_seqs = []
pka_values_seqs = []

for prot_seq in prot_seqs:
    amino_acids_info_seq_df = amino_acids_info_df.loc[list(prot_seq)]
    type_seqs.append("".join(amino_acids_info_seq_df["type_short"].tolist()))
    is_hydroxylic_seqs.append("".join(amino_acids_info_seq_df["is_hydroxylic"].tolist()))
    contains_sulphur_seqs.append("".join(amino_acids_info_seq_df["contains_sulphur"].tolist()))
    amidic_seqs.append("".join(amino_acids_info_seq_df["amidic"].tolist()))
    pka_values_seqs.append(amino_acids_info_df[["alpha_COOH", "alpha_NH+3", "RH_or_RH+"]]
                          .to_numpy().tolist())
```

```python
with open("../kaggle/input/amino_acids_info/type_sequences.json", "w+") as f:
    json.dump(type_seqs, f)
with open("../kaggle/input/amino_acids_info/is_hydroxylic_sequences.json", "w+") as f:
    json.dump(is_hydroxylic_seqs, f)
with open("../kaggle/input/amino_acids_info/contains_sulphur_sequences.json", "w+") as f:
    json.dump(contains_sulphur_seqs, f)
with open("../kaggle/input/amino_acids_info/amidic_sequences.json", "w+") as f:
    json.dump(amidic_seqs, f)
with open("../kaggle/input/amino_acids_info/pka_values_sequences.json", "w+") as f:
    json.dump(pka_values_seqs, f)
```
